```python
import matplotlib
%matplotlib inline

import os
nxg_path = os.environ.get('NXG_PATH')
print(nxg_path)

import sys
sys.path.append(nxg_path + 'core')
sys.path.append(nxg_path + 'configuration-setup')

from learningModule import DataConfiguration, CreateTrainNN

epochs = 30
dynamics = 'OtherBenchC4'
dataObject = DataConfiguration(dynamics=dynamics, dimensions=3, gradient_run=True, sensitivity='Fwd')
dataObject.setSteps(250)
dataObject.setSamples(40)
# dataObject.setTimeStep(0.01)
dataObject.setLowerBound([0.3, 0.2, 0.3])
dataObject.setUpperBound([0.5, 0.4, 0.5])
dataObject.generateTrajectories(scaling=0.01)
dataObject.showTrajectories(xindex=0, yindex=1)
dataObject.createData(jumps=[1,2,5,7,11,13,17,19])

nnObject1 = CreateTrainNN(dynamics=dynamics, dnn_rbf='RBF')
nnObject1.createInputOutput(data_object = dataObject, inp_vars=['x','xp','v','t'], out_vars=['vp'])
nnObject1.setEpochs(epochs)
nnObject1.trainTestNN(optim='SGD', loss_fn='mae', layers=1, neurons=512)
```
